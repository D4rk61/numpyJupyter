```python
import numpy as np 
import matplotlib.pyplot as plt
```

# Graficando Funciones con Python

```python
N = 100
x = np.linspace(-10,10, num=N)
def f(x):
    return np.cos(x)

y = f(x)
```

```python
plt.plot(x,y)
```

```python
# Ejemplo de una funcion exponencial 
def f(x):
    return np.exp(x)

y = f(x)

plt.plot(x,y)
```

```python
# Programando logaritmos en python

def f(x):
    return np.log2(x)

# Definiendo una nueva x, que sea mayor a 0 
x = np.linspace(0.001, 256, num=3)
plt.plot(x,f(x))
```

```python
# Funcion Seccionada 
def H(x):
    if x < 0:
        return 0
    elif x >= 0:
        return 1 

N=1000 
x = np.linspace(-10, 10, num=N)
y = H(x)
 
```

```python
# Funciones compuestas o recursivas
N = 1000
x = np.linspace(-10, 10, num=N)
def g(x):
    return x**2

def f(x):
    return np.sin(x)

y = g(x)
f_o_g = f(y)

plt.plot(x,f_o_g)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# crear un conjunto
conjunto = {(1, 2), (3, 4), (5, 6)}

# extraer los valores x e y del conjunto
x_vals, y_vals = zip(*conjunto)

# crear un objeto de figura y eje
fig, ax = plt.subplots()

# trazar los puntos en un gráfico
ax.scatter(x_vals, y_vals)

# crear un círculo para el conjunto
circulo = patches.Circle((2, 4), radius=2, facecolor='none', edgecolor='black', alpha=0.5)

# agregar el círculo al eje actual
ax.add_patch(circulo)

# mostrar el gráfico
plt.show()
```
